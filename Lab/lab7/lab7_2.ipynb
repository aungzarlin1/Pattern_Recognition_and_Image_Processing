```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('../otputs/1000_rows_house_dataset.csv')
df.info()
```

```python
def plot(data, colunm):
    if data[colunm].dtype == 'bool':
        data[colunm].value_counts().plot(kind='bar')
    elif data[colunm].dtype == 'object':
        sns.histplot(data=data, y=colunm)
    else:
        sns.histplot(data=data, x=colunm)
    plt.show()
```

```python
for colunms in df.columns:
    plot(df, colunms)
```

```python
obj = []
no_obj = []

for columns in df.columns:
    if df[columns].dtype == 'object':
        obj.append(columns)
    else:
        no_obj.append(columns)
    
print(obj)
print(no_obj)
    
```

```python
def object_plot(data, columns):
    sns.histplot(data=data, y=columns)


def plot(data, colunm):
    if data[colunm].dtype == 'bool':
        data[colunm].value_counts().plot(kind='bar')
    else:
        sns.histplot(data=data, x=colunm)
```

```python
plt.rcParams["figure.figsize"] = (20,5)
for i in range(0, len(no_obj)):
    plt.subplot(3, 6, i+1)
    plot(df, no_obj[i])
plt.show()

for i in range(0, len(obj)):
    plt.subplot(1, 3, i+1)
    object_plot(df, obj[i])
plt.show()
```
