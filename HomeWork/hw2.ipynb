```python
import tensorflow as tf
import numpy as np
```

```python
raw_data = np.random.normal(10, 1, 100)
```

## Tensorflow Version 1

```python
with tf.compat.v1.Session() as sess:
    alpha = tf.constant(0.05)
    curr_value = tf.compat.v1.placeholder(tf.float32)
    prev_avg = tf.Variable(0.)
    update_avg = alpha * curr_value + (1 - alpha) * prev_avg
    
    init = tf.compat.v1.global_variables_initializer()
    
    sess.run(init)
    
    for i in range(len(raw_data)):
        curr_avg = sess.run(update_avg, feed_dict={curr_value:raw_data[i]})
        sess.run(tf.compat.v1.assign(prev_avg, curr_avg))
        print(raw_data[i], curr_avg)
```

## Tensorflow Version 2

```python
alpha = tf.constant(0.05)
prev_avg = tf.Variable(0.)
for i in range(len(raw_data)):
    update_avg = alpha * raw_data[i] + (1 - alpha) * prev_avg
    prev_avg = update_avg
    print(raw_data[i], update_avg)
#     print(raw_data[i], float(update_avg)) # print only value
```
